##  Lab 1/Part 2/ 1. Infulenza National Summary
Data file: WHO_NREVSS_Clinical_Labs_new.csv(as on 3/8/2019)

```r
#Part 2.1
# Positive tested

library(ggplot2)
#install.packages("reshape",dependencies=T)
library(reshape)
library(reshape2)
# from ggplot2 import *

data_positive_tested <- read.csv("C:/Users/shrey/Documents/Spring 2019/Data Intensive Computing CSE 587/Lab1/Part2/1. Infulenza National Summary/WHO_NREVSS_Clinical_Labs_new.csv", 
                       skip=1, header = TRUE, stringsAsFactors = FALSE)
Week <- data_positive_tested$WEEK
year <- data_positive_tested$YEAR

weekdata <- paste(year,Week,sep="-")
# weekdata
# head(Week)
stackData <- c(data_positive_tested$TOTAL.A,data_positive_tested$TOTAL.B)
#colorr <- c()
percentPos <- data_positive_tested$PERCENT.POSITIVE
stackDFrame <- data.frame(weekdata,stackData,percentPos)

# stackDFrame

colro <- c(rep("A",22),(rep("B",22)))

#mm <- melt(data_positive_tested, id = 'data_positive_tested$WEEK')
ggplot(stackDFrame)+geom_bar(aes(factor(weekdata),stackData,fill=colro),stat="identity")+labs(x='WEEK',y='NUMBER OF SPECIMENS')+scale_y_continuous(sec.axis=~./400) +
scale_fill_manual(values=c("#ffc61a","dark green"), name="")+

theme(axis.text.x = element_text(angle = 60, hjust = 1))+
labs(title="Influenza Positive Tests Reported to CDC by U.S. Clinical Laboratories,
\n                           National Summary, 2018-2019 Season")+
geom_line(aes(as.numeric(factor(weekdata)),y=rep((data_positive_tested$PERCENT.POSITIVE)*400,2), color='1'),stat="identity" ,data = stackDFrame,size=0.65)+
geom_line(aes(as.numeric(factor(weekdata)),y=rep((data_positive_tested$PERCENT.B)*450,2), color='2'),stat="identity" ,data = stackDFrame,linetype="dashed")+
geom_line(aes(as.numeric(factor(weekdata)),y=rep((data_positive_tested$PERCENT.A)*400,2), color='3'),stat="identity" ,data = stackDFrame,linetype="dashed")+
scale_color_manual("",label=c("% Percent Positive" , "% Percent B","% Percent A"),values=c("1"="black","2"="dark green","3"="yellow4"))


```

## Lab 1/Part 2/ 5. Pediatric Deaths
Data file: Weekly.csv

```r
#2.5 Pediatric Deaths
library(ggplot2)
library(stringr)
library(reshape2)
ped<-read.csv("C:/Users/shrey/Documents/Spring 2019/Data Intensive Computing CSE 587/Lab1/Part2/5. Pediatric Deaths/Weekly.csv",stringsAsFactors=FALSE)

ped<-subset(ped,select=c('WEEK.NUMBER','TYPE','COUNT'))

break4me<-ped$WEEK.NUMBER
fill_x<-function(strt,.offset=0,.mod=12,.fill=""){
    log<-as.logical((1:length(strt)-1+.offset)%%.mod)
    strt[log]<-.fill
    strt
}
ggplot()+
  geom_bar(mapping=aes(x=ped$WEEK.NUMBER,y=ped$COUNT,fill=ped$TYPE),stat="identity",color='black') +

  xlab("Week of Death") + 
  ylab("Number of Deaths")+
scale_fill_manual("",values=c("CURRENT WEEK DEATHS"="cyan","PREVIOUS WEEK DEATHS"="dark green"))+

labs(title="Number of Influenza-Associated Pediatric Deaths,\nby Week of Death: 2015-2016 season to present",size=10)+
scale_y_continuous(breaks=seq(0,30,by=5),limits=c(0,30),expand=c(0,0))+ theme_classic()+
theme(plot.title=element_text(hjust=0.35),
     axis.text.x=element_text(angle=90,hjust=1),
     legend.position="bottom",legend.box.background=element_rect(colour="black",size=1.5))+
scale_x_discrete(name="Week of Death",breaks=break4me,labels=fill_x(break4me,.mod=12))+

annotate("text",x=25,y=26,label="2015-2016")+
annotate("text",x=70,y=26,label="2016-2017")+
annotate("text",x=125,y=26,label="2017-2018")+
annotate("text",x=165,y=26,label="2018-2019")+
annotate("text",x=25,y=24,label="Number of Deaths\nReported=94",size=3)+
annotate("text",x=70,y=24,label="Number of Deaths\nReported=110",size=3)+
annotate("text",x=125,y=24,label="Number of Deaths\nReported=185",size=3)+
annotate("text",x=165,y=24,label="Number of Deaths\nReported=41",size=3)
```
